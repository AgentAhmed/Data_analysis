### Joins

```python
# Group by 'brand' and calculate the average mileage for each brand
avg_mileage = df.groupby('brand')['mileage'].mean().reset_index()

# Rename the 'mileage' column to 'avg_mileage' for clarity
avg_mileage.rename(columns={'mileage': 'avg_mileage'}, inplace=True)
```

```python
avg_mileage
```

```python
high_km_driven_cars = df[df['km_driven']>=20000]
high_km_driven_cars
```

Removal of outliers

```python
df1 = df.copy()
df1 = df[['vehicle_age','km_driven']]
for i in list(df1.columns):
    Q1 = df1[i].quantile(0.25)
    Q3 = df1[i].quantile(0.75)
    IQR = Q3 - Q1 
    df1 = df1[df1[i]<= (Q3+(1.5*IQR))]
    df1 = df1[df1[i]>= (Q3-(1.5*IQR))]
    df1 = df1.reset_index(drop=True)

print(df1.head())    
print("Before removal" , df.shape[0])
print("After removal", df1.shape[0])
```
